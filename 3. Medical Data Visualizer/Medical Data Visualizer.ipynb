```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('medical_examination.csv')
df.head()
```

```python
df['overweight'] = (df['weight']/((df['height']*0.01)**2)).apply(lambda x: 1 if x>25 else 0)
```

```python
df['cholesterol'] = (df['cholesterol']).apply(lambda x: 1 if x>1 else 0)
df['gluc'] = (df['gluc']).apply(lambda x: 1 if x>1 else 0)
df.head()
```

```python
value_vars = ['active', 'alco', 'cholesterol', 'gluc', 'overweight', 'smoke']
df_cat = pd.melt(df, id_vars=['cardio'], value_vars=value_vars)
```

```python
plot = sns.catplot(data=df_cat, kind="count",  x="variable", hue="value", col="cardio").set_axis_labels('variable', 'total')
```

```python
plot.fig.savefig('catplot.png')
```

```python
df_heat = df[(df['ap_lo']<=df['ap_hi']) & 
            (df['height']>=df['height'].quantile(0.025)) & 
            (df['height']<=df['height'].quantile(0.975)) &
            (df['weight']>=df['weight'].quantile(0.025)) &
            (df['weight']<=df['weight'].quantile(0.975))]
```

```python
corr = df_heat.corr()
corr
```

```python
mask = np.triu(corr)
```

```python
fig, ax = plt.subplots(figsize=(9,9))
sns.heatmap(corr, vmin=0, vmax=0.25, annot=True, fmt='.1f', linewidths=0, square=True, mask=mask)
```

```python
fig.savefig('heatmap.png')
```
