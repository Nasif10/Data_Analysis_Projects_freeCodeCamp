```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
df = pd.read_csv('epa-sea-level.csv')
df.head()
```

```python
x = df['Year']
y = df['CSIRO Adjusted Sea Level']
```

```python
plt.scatter(x, y)
```

```python
plt.scatter(x, y)

last_year = df["Year"].max()
temp_year = pd.Series([x for x in range(last_year+1, 2050+1)])
new_year = pd.concat([df["Year"], temp_year])
res = linregress(x,y)
plt.plot(new_year, res.slope*new_year + res.intercept, 'r')

df2 = df[df['Year'] >= 2000]
x2 = df2['Year']
y2 = df2['CSIRO Adjusted Sea Level']
new_year_2 = pd.Series([x for x in range(2000, 2050+1)])
res2 = linregress(x2,y2)
plt.plot(new_year_2, res2.slope*new_year_2 + res2.intercept, 'g')

plt.grid()
plt.title("Rise in Sea Level")
plt.xlabel("Year")
plt.ylabel("Sea Level (inches)")
```
